```python
import gym
import torch

def run_episode(env, weight):
    state = env.reset()[0]
    total_reward = 0
    is_done = False
    timestep_max = 200
    while not is_done:
        state = torch.from_numpy(state).float()
        action = int(torch.argmax(torch.matmul(state, weight)).item())
        state, reward, is_done, is_truncated, _ = env.step(action)
        total_reward += reward
        if timestep_max < 1:
            break
        timestep_max -= 1
    return total_reward

env = gym.make('CartPole-v0')
n_state = env.observation_space.shape[0]
n_action = env.action_space.n

n_episode = 1000
best_total_reward = 0
best_weight = torch.rand(n_state, n_action)
total_rewards = []
noise_scale = 0.01
for episode in range(n_episode):
    weight = best_weight + noise_scale * torch.rand(n_state, n_action)
    total_reward = run_episode(env, weight)
    if total_reward >= best_total_reward:
        best_total_reward = total_reward
        best_weight = weight
    total_rewards.append(total_reward)
    print('Episode {}: {}'.format(episode + 1, total_reward))

print('Average total reward over {} episode: {}'.format(n_episode, sum(total_rewards) / n_episode))
```
