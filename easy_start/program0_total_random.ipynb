```python
#  https://github.com/openai/gym/wiki/Table-of-environments
# Random search policy

import torch
import gym

env = gym.make("CartPole-v0")
n_episode = 1000
total_rewards = []
print('Episode', 'Total reward')
for episode in range(n_episode):
    total_reward = 0
    is_done = False
    env.reset()
    while not is_done:
        action = env.action_space.sample()
        state, reward, is_done, is_truncated, _ = env.step(action)
        total_reward += reward
    total_rewards.append(total_reward)
    print(episode+1, total_reward)
print("average total reward:", sum(total_rewards)/n_episode)
```
